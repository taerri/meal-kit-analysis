```python
import pandas as pd
from collections import Counter, OrderedDict
```

```python
ddff=pd.read_csv("밀키트 뉴스데이터.csv")
ddff.head()
```

```python
keyword=ddff["키워드"]
type(keyword)
```

```python
nouns = []
for i in keyword.tolist():
    nouns.append(i.replace(',', ' '))
```

```python
nouns = ' '.join(nouns).split()
```

```python
nouns[:10]
```

# **단어 교체**

```python
#단어 교체
result_list=[]

for i in nouns:
    if i == '11번':
        i=i.replace('11번','11번가')
    if i=='인공지능':
        i=i.replace('인공지능','AI')
    result_list.append(i)

print(result_list)
```

# **불용어 처리**

```python
df=pd.DataFrame(result_list)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import defaultdict
from sklearn.decomposition import LatentDirichletAllocation
import pickle
```

```python
stopword1 = pd.read_csv("./불용어 처리.txt",header=None,sep="\t")
stopword1 = list(stopword1[0])
```

```python
# 직접 제거
stopword2 = pd.read_excel("단어 삭제.xlsx")
stopword2.columns =["단어","불용어여부"]
stopword2 = stopword2[stopword2.불용어여부 == "x"]
stopword2 = list(stopword2["단어"])
```

```python
stopword = list(set(stopword1 + stopword2 ))
```

```python
result= [x for x in result_list if x not in stopword]
print("불용어 개수 :",len(result))
```

```python
#단어 카운트
counts = Counter(result)
wordcount=counts.most_common()
wordcount
```

```python
df=pd.DataFrame(wordcount)
```

```python
df.to_csv('전처리된 단어.csv')
```
