```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
df=pd.read_csv('mealkit_count.csv')
df.columns=['word','count']
df=df.iloc[:500]
df.tail()
```

```python
wordList=[]
for row in df.itertuples(index=False):
    wordList.append((str(row.word),row.count))
wordList[:10]
```

```python
#wordcloud 생성
wordcloud=WordCloud(font_path='C:Windows/Fonts/NanumGothic.ttf',
                   background_color='white',
                   width=800,
                   height=600)

print(dict(wordList))

cloud = wordcloud.generate_from_frequencies(dict(wordList))
plt.figure(figsize=(10,8))
plt.axis('off')
plt.imshow(cloud)
plt.show()
```
